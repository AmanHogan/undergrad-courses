```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3, 5,6,7,4])
y = np.array([1,1,1, -1,-1,-1,-1])
N = len(x)
lam = 1
alpha = .1
epochs = 100
w = np.zeros(1)
b = np.zeros(1)


for epoch in range(epochs):
    w_grad = 0
    b_grad = 0

    for i in range(N):
        hinge_loss = (max(0, 1-(np.dot(y[i], (np.dot(w,x[i]) - b)))))

        if hinge_loss > 0:
             w_grad += np.dot(-y[i], x[i])
             b_grad += -y[i]

    w_grad += lam * w
    b_grad += lam * b

    w -= alpha * w_grad 
    b -= alpha * b_grad

print(w,b)
plt.plot(x, y, 'o', color='black');
plt.plot(x, w*x+b, linestyle='solid') 
plt.show()
```
